```python
def ejercicio_billetes(cantidad):
    billetes = [100, 50, 10, 5, 1]
    for billete in billetes:
        if cantidad >= billete:
            cantidad_billete = cantidad // billete
            cantidad = cantidad % billete
            print(f"{cantidad_billete} de {billete}")
            
```

```python
def empaquetar_bombones(total_bombones):
    cajas = [50, 20, 5, 1]
    for caja in cajas:
        if total_bombones >= caja:
            cantidad_cajas = total_bombones // caja
            total_bombones = total_bombones % caja
            print(f"{cantidad_cajas} if cantidad_caja de {caja}")
bombones = (128)
empaquetar_bombones(bombones)
```

```python
df = pd.read_csv("coderhouse/Data Science I/Dropout.csv", sep=";")
df
```

## Gráfico de barras

```python

df["Target"].value_counts().plot(kind="bar")

plt.title("Cantidad de alumnos por estado")
plt.xlabel("Target")
plt.ylabel("Cantidad")
plt.xticks(
    [0,1,2],
    ["Graduated","Dropout", "Enrolled"]
)
plt.show()
```

```python
df["Gender"].value_counts().plot(kind="bar")
plt.title("Barras x Género")
plt.xlabel("Género")
plt.ylabel("Cantidad")
plt.xticks(
    [0,1,],
    ["Male","Female"]
)
plt.show()
```

```python
df["Scholarship holder"].value_counts().plot(kind="bar")
plt.title("Barras x becas")
plt.xlabel("Beca")
plt.ylabel("Cantidad")
plt.xticks(
    [0,1,],
    ["si","no"]
)
plt.show()
```

## Histogramas

```python
plt.hist(df["Age at enrollment"], bins=20)

plt.title("Distribución de edades")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")

plt.show()
```

```python
plt.hist(df["Admission grade"], bins=30)

plt.title("Notas de admisión generalizadas.")
plt.show()
```

```python
plt.hist(df["Admission grade"], bins=50)

plt.title("Notas de admisión generalizadas.")

plt.show()
```

## Gráfico de Torta (Pie Chart)

```python
df["Target"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.ylabel("")
plt.title("Distribución de alumnos")

plt.show()
```

```python
df["Gender"].replace({
    0:"Male",
    1:"Female"}
    ).value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    labels=None
)
plt.ylabel("")
plt.title("Torta x Género")
plt.legend(
   ["Hombre", "Mujer"],
    title="Estado académico",
    loc="center left",
    bbox_to_anchor=(1,0.5)
)

plt.show()
```

```python
df["Scholarship holder"].replace({
    0:"No becado",
    1:"Becado"}
    ).value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    labels=None
)
plt.ylabel("")
plt.title("Torta x beca")
plt.legend(
   ["No becado", "Becado"],
    title="Estado académico",
    loc="center left",
    bbox_to_anchor=(1,0.5)
)

plt.show()
```

## Scatterplot

### ¿Para qué sirve?
Analizar la relación entre dos variables numéricas.

```python
plt.scatter(
    df["Admission grade"],
    df["Age at enrollment"],
    alpha= 0.6
)

plt.xlabel("Nota de admisión")
plt.ylabel("Edad")

plt.show()
```

```python
plt.scatter(
    df["Admission grade"],
    df["Age at enrollment"],
    alpha= 0.6,
)

plt.xlabel("Nota de admisión")
plt.ylabel("Edad")

plt.show()
```

```python
plt.scatter(
    df["Unemployment rate"],
    df["GDP"],
    alpha =0.6
)
plt.title("Desempleo vs GDP")
plt.xlabel("Tasa de desempleo")
plt.ylabel("GDP")

plt.show()
```
